```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.graphics.gofplots as plots
#导入pyplot模块
import matplotlib.pyplot as plt

#使用中文字体
import matplotlib as mpl
mpl.rcParams['font.family']=['Microsoft Yahei', 'sans-serif']
mpl.rcParams['axes.unicode_minus']=False

#使用svg格式，避免图形模糊
#自动显示图形
%matplotlib inline
%config InlineBackend.figure_format="svg"
```

```python
# 数据准备
#s1来自均匀分布
s1=np.random.uniform(0,100,1000)
#s2来自正态分布
s2=np.random.normal(0,1,1000)
```

```python
plots.qqplot(s1,stats.norm,fit=True,line="45")
plt.show()
```

```python
plots.qqplot(s2,stats.norm,fit=True,line="45")
plt.show()
```
