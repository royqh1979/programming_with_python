```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl

#使用中文字体
mpl.rcParams['font.family']=['Microsoft Yahei','sans-serif']
mpl.rcParams['axes.unicode_minus']=False

#使用svg格式，避免图形模糊
%matplotlib inline
%config InlineBackend.figure_format="svg"
```

```python
# lognormal分布
from sklearn.preprocessing import QuantileTransformer

s1 = np.random.lognormal(0,1,5000)
s2 = QuantileTransformer().fit_transform(s1.reshape(-1,1))

# 绘制直方图
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.hist(s1,bins=100)
ax2.hist(s2,bins=100)

plt.show()
```

```python
# 卡方分布
from sklearn.preprocessing import QuantileTransformer

s1 = np.random.chisquare(10,5000)
s2 = QuantileTransformer().fit_transform(s1.reshape(-1,1))

# 绘制直方图
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.hist(s1,bins=100)
ax2.hist(s2,bins=100)

plt.show()
```

```python
# weibull分布
from sklearn.preprocessing import QuantileTransformer

s1 = np.random.weibull(1,5000)
s2 = QuantileTransformer().fit_transform(s1.reshape(-1,1))

# 绘制直方图
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.hist(s1,bins=100)
ax2.hist(s2,bins=100)

plt.show()
```

```python
# 均匀分布
from sklearn.preprocessing import QuantileTransformer

s1 = np.random.uniform(1,2,5000)
s2 = QuantileTransformer().fit_transform(s1.reshape(-1,1))

# 绘制直方图
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.hist(s1,bins=100)
ax2.hist(s2,bins=100)

plt.show()
```

```python
# 双峰分布
from sklearn.preprocessing import QuantileTransformer

s1 = np.concatenate([np.random.normal(5,1,2500), np.random.normal(10,1,2500)])
s2 = QuantileTransformer().fit_transform(s1.reshape(-1,1))

# 绘制直方图
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
ax1.hist(s1,bins=100)
ax2.hist(s2,bins=100)

plt.show()
```
