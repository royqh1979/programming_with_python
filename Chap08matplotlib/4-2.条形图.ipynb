```python
import numpy as np
#导入pyplot模块
import matplotlib.pyplot as plt

#使用中文字体
import matplotlib as mpl
mpl.rcParams['font.family']=['Microsoft Yahei', 'sans-serif']
mpl.rcParams['axes.unicode_minus']=False

#使用svg格式，避免图形模糊
#自动显示图形
%matplotlib inline
%config InlineBackend.figure_format="svg"
```

```python
#设置默认颜色和线型
from cycler import cycler
mpl.rcParams['axes.prop_cycle'] = (
    cycler(color=['steelblue', 'red', 'forestgreen', 'darkorange']) 
    + cycler(linestyle=['-', '--', '-.', ':']))
```

```python
#数据准备
colleges=['文学院','生物学院','信息学院','园林学院','水保学院','经管学院','机械学院']
numbers1 =  [35,67,43,92,21,100,77]
numbers2 =  [51,69,32,96,37,87,81]
x=np.arange(len(colleges))

#绘图
rects = plt.bar(x-0.2,numbers1,0.35,color='darkred',label='男')
plt.bar_label(rects)
rects = plt.bar(x+0.2,numbers2,0.35,color='darkgreen',label='女')
plt.bar_label(rects)
plt.xticks(x,colleges)
plt.legend(loc='upper left')
plt.show()
```
