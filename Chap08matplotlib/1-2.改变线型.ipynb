```python
import numpy as np
#导入pyplot模块
import matplotlib.pyplot as plt

#使用中文字体
import matplotlib as mpl
mpl.rcParams['font.family']=['Microsoft Yahei', 'sans-serif']
mpl.rcParams['axes.unicode_minus']=False

#使用svg格式，避免图形模糊
#自动显示图形
%matplotlib inline
%config InlineBackend.figure_format="svg"
```

```python
#设置默认颜色和线型
from cycler import cycler
mpl.rcParams['axes.prop_cycle'] = (
    cycler(color=['steelblue', 'red', 'forestgreen', 'darkorange']) 
    + cycler(linestyle=['-', '--', '-.', ':']))
```

```python
#数据准备
x = np.linspace(0, 2*np.pi, 100)
y = np.cos(x)
y2 = np.sin(x)

#绘图
plt.plot(x,y,color='green', linestyle=':', label='余弦曲线')
plt.plot(x,y2,color='red', linestyle='--',label='正弦曲线')
plt.title("正弦和余弦图像")
plt.legend()
plt.show()
```
