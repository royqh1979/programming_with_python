```python
import numpy as np
import pandas as pd

df = pd.read_csv("学生成绩.csv",encoding="GBK")

df.head()
```

```python
#查看籍贯列的类型
df["籍贯"].dtype
```

```python
#转换为分类型
df["籍贯"]=df["籍贯"].astype("category")
```

```python
#查看籍贯列的类型
df["籍贯"].dtype
```

```python
#0-1独热编码
pd.get_dummies(df["籍贯"])
```

```python
#序数编码
pd.factorize(df["籍贯"])
```

```python
#获取所有不同的类别
df["籍贯"].dtype.categories
```
